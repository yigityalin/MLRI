```python
from pathlib import Path
import sys

root_dir = Path().resolve().parent.parent.parent.as_posix()
if root_dir not in sys.path:
    sys.path.append(root_dir)
del root_dir

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import efficientnet_v2
import tensorflow as tf

from utils import config, datasets, models
```

```python
MODEL_NAME = 'data_aug-no_base_model-5_layer_conv-5_layer_dense'
EPOCHS = 200
```

```python
train_dataset, validation_dataset, test_dataset = datasets.load_dataset()
```

```python
data_augmentation = keras.Sequential([
    layers.Rescaling(1. / 255),
    layers.RandomFlip(),
    layers.RandomRotation(factor=0.25),
])
```

```python
top = keras.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), padding='same',
                  activation=keras.activations.relu,
                  kernel_initializer='he_normal'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(filters=32, kernel_size=(3, 3), padding='same',
                  activation=keras.activations.relu,
                  kernel_initializer='he_normal'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.2),
    layers.Conv2D(filters=64, kernel_size=(3, 3), padding='same',
                  activation=keras.activations.relu,
                  kernel_initializer='he_normal'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(filters=128, kernel_size=(3, 3), padding='same',
                  activation=keras.activations.relu,
                  kernel_initializer='he_normal'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),
    layers.Conv2D(filters=256, kernel_size=(3, 3), padding='same',
                  activation=keras.activations.relu,
                  kernel_initializer='he_normal'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dense(1024, activation=keras.activations.relu,
                 kernel_initializer='he_normal'),
    layers.Dense(1024, activation=keras.activations.relu,
                 kernel_initializer='he_normal'),
    layers.Dense(512, activation=keras.activations.relu,
                 kernel_initializer='he_normal'),
    layers.Dense(256, activation=keras.activations.relu,
                 kernel_initializer='he_normal'),
    layers.Dense(4, kernel_initializer='he_normal',
                 activation=keras.activations.softmax)
])
```

```python
model = models.create_model(
    model_name=MODEL_NAME,
    preprocessing_layers=data_augmentation,
    base_model=None,
    top_layers=top,
)
```

```python
history = models.fit_model(
    model,
    train_data=train_dataset,
    validation_data=validation_dataset,
    epochs=EPOCHS
)
```
